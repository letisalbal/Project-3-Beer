```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings


%matplotlib inline
```

```python
from opencage.geocoder import OpenCageGeocode
from pprint import pprint
```

```python
key = 'be191a586a264f18b3fd3466fafa52ea'
geocoder = OpenCageGeocode(key)
```

```python
results = geocoder.reverse_geocode(44.8303087, -0.5761911)
pprint(results)
```

```python
file = "data/craft-cans/breweries.csv"
brewery_df = pd.read_csv(file)
brewery_df
```

```python
list_lat = []
list_long = []
```

```python


for index, row in brewery_df.iterrows():
    
    city = row['city']
    state = row ['state']
    query = str(city)+','+str(state)
    
    results = geocoder.geocode(query)
    lat = results[0]['geometry']['lat']
    long = results[0]['geometry']['lng']
    
    list_lat.append(lat)
    list_long.append(long)
    

```

```python
brewery_df['lat'] = list_lat
brewery_df['lon'] = list_long
```

```python
brewery_group = brewery_df.groupby('state').count()
brewery_group
```

```python
brewery_group = brewery_df.groupby('city').count()
brewery_group
```
